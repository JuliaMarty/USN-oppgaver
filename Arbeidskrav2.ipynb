```python
# Oppg 3. Lag et program med en funksjon som regner om fra grader til radianer.
# Radiantallet til vinkelen regnes s√• ut ved f√∏lgende formel: v_rad = v_grad*np.pi/180
# Resultatet v_rad skrives til skjerm med passende tekst og verdi.
import numpy as np

def Grad_to_Radian (grad) :
    """  funksjon som regner om fra grader til radianer"""
    rad = grad * np.pi/180
    return (rad)
    
v_grad = float(input('Skriv inn gradtallet:' ))
v_rad = Grad_to_Radian (v_grad)
print (round (v_grad, 2), " grader = ", round (v_rad, 2), " radianer")

```

```python
## Oppg 5) Lag et program med en funksjon som tar a og b som inn-argumenter og som s√•
# regner ut arealet og ¬´ytre¬ª omkrets til en figur satt sammen av en rettvinklet trekant og en
# halvsirkel, se figuren under. Med ¬´ytre¬ª omkrets menes samlet lengde av de sorte strekene.
# Funksjonen skal returnere arealet og ¬´ytre¬ª omkrets, som s√• skrives til skjerm med passende
# tekst. 

import numpy as np

def Areal_ytre_omkrets (a, b):
#""  funksjon som regner ut areal og ytre omkrets av et figur"""
    radius = a/2 #finne radius ut av diameter
    Areal = np.pi*radius**2/2 + b * a/2 # summen av halvsyrkelareal og trekantareal
    hypotenus = np.sqrt( b**2 + a**2) # finne ukjen side (hypotenus) i rettvinklet trekant med Pytagoras
    Omkrets = b + hypotenus + np.pi*radius # summere alle tre sider i figuren
    return (Areal, Omkrets)

a = float (input ( " oppgi a i figuren  i cm "))
b = float (input ( " oppgi b i figuren i cm " ))

Areal, Omkrets = Areal_ytre_omkrets (a, b) # call to function
print (f" Arealet til figuren er ", np.around(Areal,2), "cm2")
print (f" Omkretsen til figuren er ", np.around(Omkrets,2), "cm")
           
```
